**Importing the Dependencies**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
import warnings
warnings.filterwarnings('ignore')
```

**Data Collection & Processing**

```python
# load the data from csv file to Pandas DataFrame
titanic_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data sets/Titanic data_train.csv')
```

**Handling the Missing values**

```python
# drop the "Cabin" column from the dataframe
titanic_data = titanic_data.drop(columns='Cabin', axis=1)
```

Since Cabin column is of object datatype/categorical data and it can be neglected, so we dropped that column.

```python
# replacing the missing values in "Age" column with mean value
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

In Age column we are filling the missing values with mean age value.

Inplace = True --> means that the changed values with mean Age values will not change in the orginal dataset.

```python
# finding the mode value of "Embarked" column
print(titanic_data['Embarked'].mode())
```

```python
print(titanic_data['Embarked'].mode()[0])
```

```python
# replacing the missing values in "Embarked" column with mode value
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

Since Embarked column is of categorical data. So from the data we will find the most repeated one and we will replaced missing values with that most repeated one using mode function.
 
0 --> represents the index of the S=Southampton

```python
# check the number of missing values in each column
titanic_data.isnull().sum()
```

**Encoding the Categorical Columns**

```python
# converting categorical Columns
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

**Separating features & Target**

```python
X = titanic_data.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = titanic_data['Survived']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting the data into training data & Test data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training with Logistic Regression**

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

**Model Evaluation**

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```

Confusion Matrix

```python
# Confusion matrix for Logistic regression
# Confusion matrix can be further seen to understand true positive, false positive, true negative, and false negative.

Y_prediction = model.predict(X_test)
```

```python
pd.crosstab(Y_test, Y_prediction, rownames=['True'], colnames=['Predicted'], margins=True)
```
